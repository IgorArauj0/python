```python
#Estrutura de controle = Seleção e Interação
a = 'internet'
lista = ['internet', 'rede', 'wifi', 'cabeada']

if a in lista:
    print('Acesso a Internet Disponível')
else: ('Não possui acesso a Internet')    
```

```python
#Interação
for text in lista:
    print(text)
```

```python
b = 2   
#enquanto b=2 for menor que 10 seja impresso na tela o texto abaixo.
while b < 10:
    print(b, 'O número é menor que 10')
    b += 1 #Importante para não entrarmos em um Loop Infinito
```
